```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

## Data Cleaning

### Our first dataset is on subway runtimes, also informing about the train line stops, direction, speed, and schedule. 

```python
#Read in subway data
subway_raw = pd.read_csv('subway.csv')
subway_raw.head()
```

Originally we wanted to group the subway data by day because we saw a column in a month/day/year format, but after looking closer we found that we only have the month to go off of. Thus, we will aggregate all the datasets on the month to find relationships.

```python
subway = subway_raw.drop(['Stop Path ID', '25th Percentile Runtime', '50th Percentile Runtime', '75th Percentile Runtime', 'Scheduled Trains'], axis=1)
subway['Year'] = subway['Month'].str[-4:].astype(int)
subway['Month'] = subway['Month'].str[:2].astype(int)

subway = subway[['Year', 'Month', 'Schedule Day Type', 'Time Period', 'Line', 'Direction',
       'Average Actual Runtime', 'Actual Trains', 'Distance', 'Average Speed',
       'Average Scheduled Runtime', 'Origin Station ID',
       'Destination Station ID', 'Number of Stops']]
subway.head()
subway.dtypes
```

```python
subway['total pred_runtime'] = subway['Actual Trains'] * subway['Average Scheduled Runtime']
subway['total act_runtime'] = subway['Actual Trains'] * subway['Average Actual Runtime']
subway.head()
```

Because we are not able to discern the different days of the week and have to aggregate by month, we are grouping by all of the columns that will give us distinct trains and the way they run. This includes whether they are running forwards or backward, as well as if they are running an express loop (less stops and less distance).

Also since we aggregated a bunch of different trains together, we are now manually recalculating the average run times and the average delays accurately.

```python
subway = subway.groupby(['Year', 'Month', 'Schedule Day Type', 'Time Period', 'Line', 'Direction', 'Distance'])\
    [['Direction', 'Actual Trains', 'Distance', 'Number of Stops', 'total pred_runtime', 'total act_runtime', 'Origin Station ID', 'Destination Station ID']]\
    .agg({
        'Actual Trains': 'sum',
        'Number of Stops': 'first',
        'total pred_runtime': 'sum',
        'total act_runtime': 'sum', 
        'Origin Station ID': 'first',
        'Destination Station ID': 'first'
    }).reset_index()
subway['avg pred_runtime'] = subway['total pred_runtime'] / subway['Actual Trains']
subway['avg act_runtime'] = subway['total act_runtime'] / subway['Actual Trains']
subway['avg delay'] = subway['avg act_runtime'] - subway['avg pred_runtime']
subway = subway.drop(['total pred_runtime', 'total act_runtime'], axis=1)

subway.head(10)
```

### Our second dataset is the MTA stations, which we will merge to get information on the boroughs that the train runs through

```python
# Read in Stations data
stations_raw = pd.read_csv('MTA_Subway_Stations.csv')
stations_raw.head()
```

We select useful information or columns, including the stop ID which we will be merging on, the borough, and whether the station is in the Manhattan Central business district.

```python
stations = stations_raw[['GTFS Stop ID', 'Borough', 'CBD']]
stations['CBD'] = stations['CBD'].astype(int)
stations.head()
```

### Now we will merge the two datasets to get more information about the location that the trains are running through

```python
# merging for origin station
subway = subway.merge(stations, how='left', left_on='Origin Station ID', right_on='GTFS Stop ID')
subway = subway.rename(columns={'Borough': 'origin_borough', 'CBD': 'origin_cbd'})
subway = subway.drop('GTFS Stop ID', axis=1)
```

```python
# merging for destination station
subway = subway.merge(stations, how='left', left_on='Destination Station ID', right_on='GTFS Stop ID')
subway = subway.rename(columns={'Borough': 'destination_borough', 'CBD': 'destination_cbd'})
subway = subway.drop('GTFS Stop ID', axis=1)
subway.head()
```

### Our third dataset has weather information to use in predicting delays.

```python
# reading in the weather dataset
weather_raw = pd.read_csv('more_weather_v2.csv')
weather_raw.head()
```

We keep the columns of interest, being time and those that may impact train delays. To clean the dataset, we drop NaN values and convert datatypes. For time, we take the year and month as integers.

```python
# drop NaNs
weather = weather_raw[['time', 'temperature_2m (°C)', 'precipitation (mm)', 'windspeed_10m (km/h)', 'winddirection_10m (°)']].dropna()

# transform time
weather['Year'] = weather['time'].str[:4].astype(int)
weather['Month'] = weather['time'].str[5:7].astype(int)

# select columns
weather = weather.drop('time', axis=1)
weather = weather[['Year', 'Month', 'temperature_2m (°C)', 'precipitation (mm)', 'windspeed_10m (km/h)', 'winddirection_10m (°)']]
weather.head()
```

Aggregating by month to match the other datasets, we take the total for precipitation while the means for temperature, wind speed, and wind direction.

```python
weather = weather.groupby(['Year', 'Month']).agg({
    'temperature_2m (°C)': 'mean',
    'precipitation (mm)': 'sum',
    'windspeed_10m (km/h)': 'mean',
    'winddirection_10m (°)': 'mean'
}).reset_index()
weather.head()
```

Now we merge the two dataframes together. We do a left merge as weather data is not the main variable for prediction and we could potentially train a model exluding it. Thus, we do not want to exclude potentially useful train information here due to missing weather data.

```python
data = pd.merge(subway, weather, on=['Year', 'Month'], how='left')
data.head()
```

We also want to consider wind direction in relation to the train direction, so we transform the degrees of wind direction to labels to match that from the subway data. Then, we define a new column that indicates if wind is in the same direction.

```python
# turning wind direction into categories
conditions = [
    (data['winddirection_10m (°)'] >= 90) & (data['winddirection_10m (°)'] < 180),   # South
    (data['winddirection_10m (°)'] >= 270) | (data['winddirection_10m (°)'] < 90)     # North
]

# Define the direction labels in the same order as conditions
directions = ['S', 'N']

# Transform column with the mapped directions
data['winddirection_10m (°)'] = np.select(conditions, directions, default='Unknown')

# Create new column for whether wind and train are in same direction
data['wind_same_direction'] = data['Direction'] == data['winddirection_10m (°)']
data = data.drop(columns=['Direction', 'winddirection_10m (°)'])
data.head()
```

As we are also only interested in the delays, we omit the negative delays where the trains arrive early.

```python
data = data[data['avg delay'] > 0]
```

To get a definite measurement of the linear relationships, we calculate Pearson's correlation coefficient.

```python
# function to properly report the pearson coefficient
d_f = data.shape[0]-2
def report_pearson(x, y):
    p = scipy.stats.pearsonr(x, y)
    return f'r({d_f}) = {round(p.statistic, 3)}, p = {round(p.pvalue, 3)}'

# apply the function across the variables
corr = []
diag = 0
for i in ['avg delay']+continuous:
    diag+=1
    for c in (['avg delay']+continuous)[:diag]:
        d = data[[i, c]].dropna()
        corr.append([i, c, report_pearson(d.iloc[:,0], d.iloc[:,1])])
    for c in (['avg delay']+continuous)[diag:]:
        corr.append([i, c, ''])

# convert to dataframe then pivot table
pd.DataFrame(corr, columns=['x', 'y', 'r']).pivot(index='x', columns='y', values='r').reindex(
    index = ['avg delay']+continuous, columns = ['avg delay']+continuous)
```

Each result is formatted: r(degrees of freedom) = Pearson coefficient, p = p-value
- degrees of freedom = test parameter related to the number of data points
- Pearson coefficient = relates to the strength of the correlation, absolute value closer to 1 is stronger, sign represents positive or negative relationship
- p-value = likelihood of the Pearson coefficient if there was no correlation, relating to how significant or valid the result is, at a significance level of 0.05 we consider the result significant if p is less than 0.05

Comparing the variables to `avg delay`, we see the variables generally have weak correlations but do significantly correlate (high p-value). Though the correlations are weak, they are expected: `Number of Stops` will slow down trains from having to stop to exchange passengers; `Distance` will slow down trains as there are more opportunities to be delayed by another factor (like more stops or construction); `windspeed_10m (km/h)` will slow down trains as more energy is needed to fight wind; `temperature_2m (°C)` will slow down trains as more heat may impact servicers and potentially train efficiency. A negative relationship with `precipitation (mm)` was not expected as it typically disrupts routine and increases delays, though the relationship is overall weak.

We again see significant correlation between `Distance` with `Number of Stops` of r=0.825 and `windspeed_10m (km/h)` with `temperature_2m (°C)` of r=0.571. We will not consider `Number of Stops` in our model because of the strong relationship as retaining both of the variables per pair may lead to multicollinearity. Multicollinearily would not impact the model results but impede the interpretability of the model weights. We retain the other variables as their relationship is not as strong.

# Model

For our model, the features we are incorporating are:
- number of stops, actual trains, time of day, weekday/weekend, origin_borough, destination_borough, line, precipitation, windspeed, temperature

We are using RandomForest instead of simple linear regression to better capture the relationships, as previously observed from EDA, there are not many strong linear correlations but there may be more complex ones.

First, we create separate features (X) and results (y) dataframe and series respectively.

```python
features = data[['Month','Time Period','Schedule Day Type', 
                 'origin_borough', 'destination_borough', 'Line','Distance',
                 'origin_cbd', 'destination_cbd',
                 'temperature_2m (°C)', 
                 'precipitation (mm)', 'windspeed_10m (km/h)', 'wind_same_direction']]
delay = data['avg delay']

features.dtypes
```

We separate these features into bins depending on what feature transformations we need to apply to them, which bases on whether they are numerical or categorical.

```python
numeric_features = [
    "Distance",
    "temperature_2m (°C)",
    "precipitation (mm)",
    "windspeed_10m (km/h)"
]

categorical_features = [
    "Time Period",
    "Schedule Day Type",
    "origin_borough",
    "destination_borough",
    'origin_cbd', 'destination_cbd',
    "Line",
    "wind_same_direction",
    "Month"
]
```

We create pipelines to transform the data in these new bins then combine then into a single column transformer.

```python
# Create pipelines
numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# Combine pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
        # ("bool", bool_transformer, bool_features),
    ]
)
```

Then, we build the final pipeline which includes the preprocessor and the regressor.

```python
model_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("regressor", RandomForestRegressor(n_estimators=100, random_state=42))
])
```

We split the data into training and test sets to be able to evaluate our model.

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, delay, test_size=0.2, random_state=42
)
```

We fit or train the model pipeline to the training data.

```python
model_pipeline.fit(X_train, y_train)
```

Make predictions on the test set

```python
y_pred = model_pipeline.predict(X_test)
```

We find the R^2 or score of this model to evaluate.

```python
model_pipeline.score(X_test, y_test)
```

We look at the most important features considered in the model below, starting from the most to least.

```python
# Get feature importances from the RandomForestRegressor
importances = model_pipeline.named_steps["regressor"].feature_importances_

# Get the transformed feature names
feature_names = (
    model_pipeline.named_steps["preprocessor"]
    .get_feature_names_out(input_features=features.columns)
)

# Create a DataFrame to display feature importances
feature_importance_df = pd.DataFrame({"Feature": feature_names, "Importance": importances})
feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

# Display feature importances
feature_importance_df.head(11)
```

Our model ultimately achieved and R^2 of 0.78, falling short of our 0.80 goal, but still delivering strong predictive performance despite the data constraints. We initially expected that averaging delay times across months would significantly lower performace, yet the model was still able to capture meaningful trends.

Our findings highlight several key takeaways regarding subway delays in New York City.
